# Padding frames and RNN Model 
## RNNs and Sequence Length 
- RNNs (Recurrent Neural Networks) process sequential data, such as text, audio, or videos, one element (e.g., frame) at a time.

- Input requirement: When batching multiple sequences together for training (for efficiency), each sequence in the batch must be the same length.

    -  With video data, each “sequence” is a video, and each “element” is a frame.

# The Challenge with Video Data
- Real-world video clips usually have variable lengths.

    - For example, one video might have 500 frames, another 800, another 314, etc.

- RNNs expect uniform sequence lengths when working in batches.

    - This is a common requirement in deep learning frameworks (like TensorFlow, PyTorch, Keras).
    
    
# How Padding is Used for RNNs
1. Padding Sequences
- Shorter videos are padded with extra (blank) frames to reach a fixed length (usually the length of the longest video in the batch, or a predefined max length).

- These padded frames are typically all zeros (black frames).

2. Masking
- Simply padding isn’t enough! We don’t want the RNN to treat the padded (fake) frames as real data.

- So, **masking** is used:

    - You provide the model with a mask indicating which frames are real and which are padded.

    - Most deep learning libraries have a way to automatically ignore the padded parts during training and loss calculation.
